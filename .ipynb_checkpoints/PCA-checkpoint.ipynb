```python
# Import relevant libraries 
import string
import numpy as np
import matplotlib.pyplot as plt
import collections
from itertools import tee, islice
from sklearn.decomposition import PCA
```

```python
# Visualize these words in a 2D space (spanned by 1st and 2nd principal components) via PCA where PCA is applied to bi-gram word-context 
# matrix from Ulysses [see below]
query_words = ['run','walk','shout','talk','speak','animal','duck','chicken','love','hate'];
```

```python
# Load input text file for analysis 
# (Ulysses by James Joyce, 1922)
txt = open("ulysses.txt","r")
```

# Text Cleaning

```python
# Remove punctuations in text
remove = dict.fromkeys(map(ord, string.punctuation))
```

```python
# Collapse tokens to lower case
txt = txt.read().translate(remove).lower()
```

```python
# Code for obtaining ngram counts with inputs of "lst" = a list and "n" = n-gram order, e.g. n=2 -> digram counts
# For instance, this bi-gram model slices given text by two adjacent words.
def ngrams(lst, n):
    tlst = lst
    while True:
        a, b = tee(tlst)
        l = tuple(islice(a, n))
        if len(l) == n:
            yield l
            next(b)
            tlst = b
        else:
            break
```

```python
# Split text into separate words
txt_split = txt.split();
def getUniqueWords(allWords) :
    uniqueWords = [] 
    for i in allWords:
        if not i in uniqueWords:
            uniqueWords.append(i)
    return uniqueWords
unique = getUniqueWords(txt_split)
# print (len(unique))
```

# Analysis

```python
# First construct a dictionary that counts the number of time any word combinations occur.
a = collections.Counter(ngrams(txt_split, 2)) # a is a dictionary structured as {(word 1, word 2): number of occurences}

# Extract the combinations that interest us
comb_interest = [] # instances where a query word is followed by a unique word #com
num_occur = [] # number of times that the combination a query word occurs before a unique word #c
for q in query_words:
    for w in unique:
        n = a[(q, w)]
        num_occur.append(n)
        comb_interest.append((q,w))
```

```python
# Step 1: Construct a word-by-context matrix C where: 
# word (rows) = query_words provided
# context (columns) = all unique words in txt_split
# C[i,j] = Number of times that word i appears before context j in txt_split
matrix_c = np.array([a, b, c, d, e, f, g, h, i, j])
print (type(matrix_c))
print (matrix_c)
```

```python
# Step 2: Apply PCA to C
pca = PCA(n_components = 2)
pca.fit(matrix_c)
#print (pca)
first_pc = pca.components_[0]
second_pc = pca.components_[1]
transformed_data = pca.transform(matrix_c)
# print (first_pc)
# print (second_pc)
# print (transformed_data)

```
